```python
import pandas as pd
```

```python
df1 = pd.read_csv('/Users/cheyen/Desktop/JEDHA/FULLSTACK/final_project/data_clean/hellowork.csv')
df2 = pd.read_csv('/Users/cheyen/Desktop/JEDHA/FULLSTACK/final_project/data_clean/linkedin_wttj_indeed.csv')

df1.head()
```

```python
df2.head()
```

```python
df1['job_description'] = df1['job_description'] + ' ' + df1['profile'].fillna('')
df1.drop(['profile', 'job_type'], axis=1, inplace=True)
```

```python
df1.rename(columns={'job_location': 'location', 'all_skills': 'skills', 'source' : 'platform'}, inplace=True)

df1.head()
```

```python
df1 = df1[["job_title", "job_company", "job_description", "skills", "location", "platform", "salary"]]
df1.head()
```

```python
print(df1.shape)
print(df2.shape)
```

```python
merged_df = pd.merge(df1, df2, on=['job_title', 'job_company', 'job_description', 'skills', 'location', 'platform'], how='outer')

print(merged_df.shape)
```

```python
merged_df
```

```python
merged_df = merged_df.drop_duplicates(subset=['job_company', 'job_description'], keep='first')
merged_df.shape
```

```python
merged_df.to_csv('jobs_list.csv', index=False)
```
