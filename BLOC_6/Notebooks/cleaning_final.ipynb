```python
df['job_description'] = df['job_description'].str.lower()

## Extraction de la localisation des offres grâce à la description
# Listes des villes principales
cities = ['marseille', 'paris', 'lyon', 'toulouse', 'montpellier', 'lille', 'rennes', 'reims', 'nice', 'île de france', 'nancy', 'brest', 'strasbourg',
'nantes', 'tourcoing', 'roubaix', 'bordeaux', 'nanterre', 'tours', 'les-clayes-sous-bois', 'écully', 'chaponnay', 'levallois-perret','saint-ouen-sur-seine',
'castelnaudary','le mans', "val-d'oise", "villeneuve-d'ascq", 'neuilly-sur-seine', 'metz', 'toulon', 'beynost', 'courbevoie', 'olivet', 'dardilly', 'loos', 
'bezons', 'dax', 'issy-les-moulineaux', 'malakoff', 'saint-jacques-de-la-lande','niort','charenton-le-pont', 'saint-herblain', 'carquefou', 'vienne']

colonne = 'job_description'

# Fonction permettant de rechercher le nom des villes dans la description
def found_cities(description):
    cities_found = []
    for city in cities:
        if city in description:
            cities_found.append(city)
    return cities_found if cities_found else np.nan

df['cities'] = df[colonne].apply(found_cities)
df['cities'] = df['cities'].str.join(', ')
```

```python
df_geo['job_title'] = df_geo['job_title'].str.lower()

# Liste des mots à filtrer
filter_words = ['data', 'donnée','données', 'donnee', 'donnees' 'developer', 'développeur', 'développeuse', 'dev', 'cloud', 'bi', 'business intelligence', 
                'analyst','analyste', 'analytics' , 'analytic' 'analyse', 'engineer', 'ingénieur', 'ml', 'mlops', 'machine learning', 'software', 'python',
                'growth marketing','ia', 'intelligence artificielle', 'deep learning', 'growth marketer','crm','product owner', 'technicien informatique',
                'scientist', 'science','statisticien', 'études statistiques', 'algorithme', 'marketing', 'ingenieur', 'digital', 'informatique', 'product manager',
                ]

# Utilisation d'une expression régulière pour créer un motif de recherche
pattern = r"(?i)" + "|".join(filter_words)

# Filtrage de la colonne "job_title"
filtered_jobs = df_geo[df_geo['job_title'].str.contains(pattern) == False]

# Affichage des résultats
new_df = pd.DataFrame({'job_title': filtered_jobs['job_title']})
pd.set_option('display.max_rows', None)

new_df.shape
```

```python
import re
def remove_emoji(string):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)

df_geo['job_title'] = df_geo['job_title'].apply(remove_emoji)
df_geo['job_description'] = df_geo['job_description'].apply(remove_emoji)
```

```python
#df_geo.to_csv('dataset_finalv2.csv', index=False)
```

```python
new_df_geo = df_geo[df_geo['location'].notnull()].copy()
new_df_geo.head()
```

```python
new_df_geo['location'] = new_df_geo['location'].astype(str).apply(lambda x: ', '.join([x, 'France']))
```

```python
import geocoder

# Extraction de la latitude des villes
def get_latitude(location):
    g = geocoder.osm(location)
    if g is not None and g.json is not None:
        return g.json.get('lat')
    return None

new_df_geo['lat'] = new_df_geo['location'].apply(get_latitude)
```

```python
#Extraction de la longitude
def get_longitude(location):
    g = geocoder.osm(location)
    if g is not None and g.json is not None:
        return g.json.get('lng')
    return None

new_df_geo['lon'] = new_df_geo['location'].apply(get_longitude)
```

```python
# Concaténer les deux DataFrames en utilisant l'index et en ajoutant uniquement les colonnes de new_df_geo qui ne sont pas présentes dans df_geo
final_df = pd.concat([df_geo, new_df_geo[columns]], axis=1, join='outer')
```

```python
#Fonction permettant d'assigner des rôles pour faciliter le tri des types d'emploi
def assign_role(x):
  
    import re
    if re.search(r'scien.que|data.scien|data.science|machine.learning|modélisation|datascientist',x):
        return "Data Scientist"
    
    elif re.search(r'architect|cloud|big.data',x):
        return "Data Architect"
    
    elif re.search(r'engineer|ing.nieur|ingénierie',x):
        return "Data Engineer"
       
    elif  re.search(r'statistiques|statisticien',x):
        return "Statisticien"
        
    elif re.search(r'busines..analyst|busines..manager|busines..develop',x):
        return "Business Analyst"
    
    elif  re.search(r'data.analys|analys..data|analyste de données|analytics|designer|bi|intelligence|power|visualisation|décisionnelle|visualization',x):
        return "Data/BI Analyst"
    #elif  re.search(r'bi|intelligence|power',x):
    #    return "BI Analyst"
    
    elif re.search(r'développeu|developpeu|développement|developer|ops|program|concepteur',x):
        return "Développeur"

    elif re.search(r'consultant|digital|change maker|expert|stratégie|amoa',x):
        return "Consultant"
    
    elif re.search(r'product.owner|product.manager|po',x):
        return "Product Owner/Manager"
    
    elif  re.search(r'technicien|technique|réseau|admin',x):
        return "Technicien/Administrateur Data"
    
    elif re.search(r'steward|coordinator|gouvernance|governance|master|gestionnaire|protection|officer|dataminer|intégrateur|flux|traitement de données|entry|support|exploitation',x):
        return "Gestionnaire Data"
    
    elif re.search(r'projet|mission|pmo',x):
        return "Chef de projet"
    
    elif re.search(r'manager|management|lead|head|responsable|directeur|chief|chef',x):
        return "Data Manager"
    
    elif re.search(r'crm|sales|marketing|marketing.digital|commercial|growth|client|affaires|communication|media|web|market',x):
        return "Pôle commercial/marketing"
    
    elif re.search(r'comptable|gestion|reporting|financ|performance|qa|m&a|jurist|account|middle.office|hr|rh|recrutement|talent|notar|actuair|économ|investissement|audit',x):
        return "Pôle finance/HR"
      
    else:
        return "Autres"
```

```python
final_df['job_class'] = final_df['job_title'].str.lower().apply(assign_role)
```
