```python
#Supprimer les lignes dupliquées 
df = df.drop_duplicates(subset=['job_description'])
df.shape
```

```python
#Supprimer les caractères indésirables
df = df.replace(r'\n', '', regex=True)
df = df.replace(r'\t', '', regex=True)
df['job_location'] = df['job_location'].str.extract(r'>(.*?)<')

#Changement de nom de la colonne "type"
df['job_type'] = df['type']
df = df.drop(['type'], axis=1)

df.head()
```

```python
#Vérification des valeurs manquantes
print('The number of missing values by columns is :\n', df.isna().sum())
```

```python
#Remplacement des valeurs manquantes de la colonne "type"
df['job_type'] = df['job_type'].fillna(df['job_type'].value_counts().index[0])
df.head()
```

La colonne "profile" est vide lorsque le profil recherché par l'entreprise est inscrit directement dans la description du poste. 

```python
df['salary'].value_counts()
```

```python
## Nettoyage de la colonne "salary"

# Suppression des caractères indésirables
df['salary'] = df['salary'].str.replace('\u202f', '')
df['salary'] = df['salary'].str.replace(' EUR par an', '')
df['salary'] = df['salary'].str.replace(' EUR par mois', '')
df['salary'] = df['salary'].str.replace(' EUR par jour', '')
df['salary'] = df['salary'].str.replace(' EUR par heure', '')

# Création de colonne min et max pour séparer les tranches de salaires 
df[['salary_min', 'salary_max']] = df['salary'].str.split('-', expand=True)
df['salary_min'] = df['salary_min'].str.replace(',','.').str.strip()
df['salary_max'] = df['salary_max'].str.replace(',','.').str.strip()

# Conversion des colonnes en float
df['salary_min'] = df['salary_min'].astype(float)
df['salary_max'] = df['salary_max'].astype(float)

df.head()
```

```python
# Mise à l'échelle des salaires mensuels, journaliers et horaires en salaire annuel
df['salary_min'] = df['salary_min'].apply(lambda x: x * 12 if x >= 1000 and x < 10000 else (x * 20 * 12 if x >= 100 and x < 1000 else (x * 7 * 20 * 12 if x >= 5 and x < 100 else x)))
df['salary_max'] = df['salary_max'].apply(lambda x: x * 12 if x >= 1000 and x < 10000 else (x * 20 * 12 if x >= 100 and x < 1000 else (x * 7 * 20 * 12 if x >= 5 and x < 100 else x)))
df['salary_max'].value_counts()
```

```python
df.describe()
```

```python
# Création d'une nouvelle colonne comprenant la moyenne entre le salaire min et le salaire max
df['salary'] = (df['salary_min'] + df['salary_max']) / 2

# Calcul du mode de la colonne 'salary'
mode = df['salary'].mode()[0]

# Remplacer les valeurs manquantes de la colonne "salary" par le mode
df['salary'].fillna(mode, inplace=True)

# Suppression des colonnes "salary_min" et "salary_max"
df = df.drop(['salary_min', 'salary_max'], axis=1)

df.head()
```

```python
# Extraction des compétences de la colonne "job_description

skills = ["Python", "SQL", "AWS", "Spark", "Azure", " R ", "Tableau", "Java", "Excel", "Scala", "Hadoop", "Power BI", "Snowflake", "Kafka", "NoSQL", "DataBricks", "Redshift",
"Git", "Oracle", "Airflow", "GCP", "Docker", "SQL Server", "SAS", "Kubernetes"]
colonne = 'job_description'

def found_skills(description):
    skills_found = []
    for skill in skills:
        if skill in description:
            skills_found.append(skill)
    return skills_found if skills_found else np.nan

df['skills'] = df[colonne].apply(found_skills)
df['skills'] = df['skills'].str.join(', ')
```

```python
# Extraction des compétences de la colonne "profile"

skills = ["Python", "SQL", "AWS", "Spark", "Azure", " R ", "Tableau", "Java", "Excel", "Scala", "Hadoop", "Power BI", "Snowflake", "Kafka", "NoSQL", "DataBricks", "Redshift",
"Git", "Oracle", "Airflow", "GCP", "Docker", "SQL Server", "SAS", "Kubernetes"]

def extract_skills(profile):
    skills_found = []
    for skill in skills:
        if skill in profile:
            skills_found.append(skill)
    return skills_found if skills_found else np.nan

df['profile'] = df['profile'].astype(str)
df['found_skills'] = df['profile'].apply(extract_skills)
df['found_skills'] = df['found_skills'].str.join(', ')
```

```python
# Concaténation des colonnes "skills" et "found_skills"
df['all_skills'] = df.apply(lambda row: ', '.join(set(str(row['skills']).split(', ') + str(row['found_skills']).split(', '))) if pd.notnull(row['skills']) and pd.notnull(row['found_skills']) else row['skills'], axis=1)

# Suppression des colonnes "skills" et "found_skills"
df = df.drop(['skills', 'found_skills'], axis=1)

df.head()
```

```python
df['job_class'] = df['job_title'].str.lower()

# Définir les correspondances pour regrouper les titres d'emploi
mappings = {
    'Data Scientist': ['scientist', 'science'],
    'Data Engineer': ['engineer', 'eng', 'ingénieur', 'ingé'],
    'Data Analyst': ['analyst', "analyse"],
    'Business Intelligence': ['business', 'business intelligence'],
    'Développeur': ['développeur', 'developer'],
    'Architecte de données': ['architecte', 'architect'],
    'Machine Learning Engineer': ['machine learning engineer', 'ml engineer', 'ml'],
    'Data Manager' : ['manager', 'management'],
    'Consultant' : ['consultant'],
    'Chargé de projet' : ['projet', 'project'],
    "Chargé d'études" : ['études', "etude", 'etudes', 'study'],
    "Technicien" : ['technicien', "tech", "tech lead"],
    }

def replace_job_title(title):
    for key, values in mappings.items():
        for value in values:
            if value in title.lower():
                return key
    return 'Autre'  # Valeur par défaut pour les titres non correspondants

# Appliquer les correspondances pour regrouper les titres d'emploi
df['job_class'] = df['job_title'].apply(replace_job_title)

df.head()
```
