### CLEANING DATA

```python
## Remove useless columns

columns_to_remove = ['id', 'idg', 'condtn', 'undergra','tuition','from','income','expnum','zipcode','career','exphappy','satis_2',
                     'length','numdat_2','you_call','them_cal','date_3','numdat_3','num_in_3','attr1_3','sinc1_3','intel1_3','fun1_3'
                     ,'amb1_3','shar1_3','attr2_3','sinc2_3','intel2_3','fun2_3','amb2_3','shar2_3','attr3_3','sinc3_3','intel3_3','fun3_3'
                     ,'amb3_3','attr4_3','sinc4_3','intel4_3','fun4_3','amb4_3','shar4_3','attr5_3','sinc5_3','intel5_3','fun5_3'
                     ,'amb5_3','attr7_3','sinc7_3','intel7_3','fun7_3','amb7_3','shar7_3']

data.drop(columns_to_remove, axis=1, inplace=True)
data.shape
```

```python
## Modify columns : field_cd, race, gender

new_fildcd = {1.0 : "Law",
                 2.0 : "Math",
                 3.0 : "Social Sc.",
                 4.0 : "Medical Sc.",
                 5.0 : "Eng.",
                 6.0 : "Journalism",
                 7.0 : "History",
                 8.0 : "Economy",
                 9.0 : "Education",
                 10.0 : "Bio.Science",
                 11.0 : "Social Work",
                 12.0 : "Undecided",
                 13.0 : "Politics",
                 14.0 : "Film",
                 15.0 : "Arts",
                 16.0 : "Languages",
                 17.0 : "Architecture",
                 18.0 : "Others"}

data['field_cd'] = data['field_cd'].replace(new_fildcd)

new_race = {1: 'Black',
            2: 'White',
            3:'Hispanic',
            4:'Asian',
            5:'Native',
            6:'Others'}

data['race'] = data['race'].replace(new_race)

new_gender = {0 : 'Female',
              1 : 'Male'}

data['gender'] = data['gender'].replace(new_gender)
```

```python
data.head()
```

```python
## Remove the duplicates raws to visualize the participants profiles

data2 = data.drop_duplicates(subset=['iid'], keep='first')
data2.shape
```

### DATA VISUALIZATION

```python
## Starting to analyze who are the participants

print(data2['gender'].value_counts())
print()
print(data2['race'].value_counts())
print()
print(data2['field_cd'].value_counts())
```

```python
# Some distributions plot to better understand our participants

fig, axes = plt.subplots(4, figsize=(20,20))


sns.histplot(data=data2, x='gender', stat='density', hue='gender', ax=axes[0])
sns.histplot(data=data2, x='age', kde=True, stat='density', ax=axes[1])
sns.histplot(data=data2, x='race', stat='density', hue='race', ax=axes[2])
sns.histplot(data=data2, x='field_cd', hue='field_cd', stat='density', ax=axes[3])
sns.set()
plt.subplots_adjust(hspace=1)

plt.show()
```

As we can see the proportion of men and women is balanced like we could imagine for a speed dating study. 

We have a normal age distribution with mostly people in their twenties. Some outliers with older people. The participants are mainly white not surprising in United States I think.

Finally we can see that Economy and Biological Science are the most studied fields.

```python
## What can be interested is to check why the participant decided to participate.

fig = px.histogram(data2, x='goal', color='gender', nbins=20)

fig.show()
```

Here is the explanation of the encoded values : 

1.0 = Seemed like a fun night out

2.0 = To meet new people

3.0 = To get a date

4.0 = Looking for a serious relationship

5.0 = To say I did it

6.0 = Others

Participants mostly want to meet new people or get the experience because it looks fun. We don't have a lot of differences between male and female except for the "3" point. As we see, men are more likely to participate in order to get a date than women.

```python
columns = ['attr1_1', 'sinc1_1', 'intel1_1', 'fun1_1', 'amb1_1', 'shar1_1']

# average of the scores given for each attribute 
radar_chart = data2.groupby('gender')[columns].mean()
radar_chart
```

"attr" correspond to participants' ratings of 6 attributes : attractive, sincerity, intelligence, fun, ambitious, shared_interest

"1_1" is the code assigned according to the question, in this case: "What are you looking for in the opposite sex ?"

As we see the scores are fairly balanced except for "attr1_1" where male have much more higher score and "amb1_1" where female have much more higher score. It is funny because this highlights the clichés that men are mostly interested in looks, and women in status.

```python
# Reset the index to bring 'gender' back as a normal column
radar_chart = radar_chart.reset_index()

# Reshape the DataFrame using pd.melt to create a 'variable' column for the attributes and a 'value' column for the ratings
radar_chart_melted = pd.melt(radar_chart, id_vars='gender', value_vars=columns, var_name='attribute', value_name='rating')
```

```python
new_attrib = {'attr1_1': 'Attractive',
            'sinc1_1': 'Sincere',
            'intel1_1':'Intelligence',
            'fun1_1':'Funny',
            'amb1_1':'Ambitious',
            'shar1_1':'Shared Int.'}

radar_chart_melted['attribute'] = radar_chart_melted['attribute'].replace(new_attrib)
radar_chart_melted
```

```python
# Create the radar chart using Plotly Express
fig = px.line_polar(radar_chart_melted, r='rating', theta='attribute', line_close=True, color='gender')

fig.update_layout(title='What participants look for in a partner')
fig.update_traces(fill='toself', textposition='top center')
fig.show()
```

What they think opposite sex look for in a partner ?

"2_1" is the code assigned according to the question, in this case: "What do you think the opposite sex look for ?"

This column gives us what people think opposite sex research and we can see that women gave a very high score to attractiveness which is pretty accurate because it was the highest score given by men. Even if the women gave a slightly too high score. We can observe the same thing for "ambition", women gave a low score like the men did previously. On the other hand, women greatly underestimate men's search for intelligence and sincerity.

Now if we look at the men's scores, we see that like women, they gave attractiveness the highest score while women place more importance on intelligence, according to the ratings given above. They completely underestimate the importance of sincerity and intelligence too and overestimate a little the importance of ambition.

```python
# Reset the index to bring 'gender' back as a normal column
radar_chart2 = radar_chart2.reset_index()

# Reshape the DataFrame using pd.melt to create a 'variable' column for the attributes and a 'value' column for the ratings
radar_chart_melted2 = pd.melt(radar_chart2, id_vars='gender', value_vars=columns2_1, var_name='attribute', value_name='rating')

new_attrib2 = {'attr2_1': 'Attractive',
            'sinc2_1': 'Sincere',
            'intel2_1':'Intelligence',
            'fun2_1':'Funny',
            'amb2_1':'Ambitious',
            'shar2_1':'Shared Int.'}

radar_chart_melted2['attribute'] = radar_chart_melted2['attribute'].replace(new_attrib2)
radar_chart_melted2
```
